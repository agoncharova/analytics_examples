## Bar Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv(r"data/bar_chart_data.csv")
df_used_cars
```

```python
plt.figure(figsize = (9,6)) # increasing the size for a better visualisation
plt.bar(x = df_used_cars["Brand"], height = df_used_cars["Cars Listings"], color = 'g')
plt.xticks(rotation = 45)
plt.title("Car Listings by Brand",fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used Car Bar.png")
```

## Pie Chart

```python
df_fuel_engine_type = pd.read_csv(r"data/pie_chart_data.csv")
```

```python
df_fuel_engine_type
```

```python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],
        labels = df_fuel_engine_type['Engine Fuel Type'].values,
        autopct = '%.2f%%', 
        textprops = {'size':'x-large','fontweight':'bold', 'rotation': '30','color': 'w'})
plt.legend()
plt.title('Cars by Engine Fuel Type', fontsize = 18, fontweight = 'bold')
plt.show()
```

## Stacked Area Chart

```python
df_fuel_engine_types = pd.read_csv(r"data/stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
colors = ["#011638","#7e2987","#ef2026"]
labels = ["Diesel","Petrol","Gas"]
sns.set_style("white")
plt.figure(figsize = (12,6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors,
              edgecolor = 'none')
plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of cars", fontsize = 13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 14, weight = 'bold')
sns.despine()
plt.show
```

## Line Chart

```python
df_spx_ftse = pd.read_csv(r"data/line_chart_data.csv")
```

```python
df_spx_ftse
```

```python
df_spx_ftse["new_date"] = pd.to_datetime(df_spx_ftse["Date"])
```

```python
df_spx_ftse
```

```python
labels = ["S&P", "FTSE 100"]
plt.figure(figsize = (20,8))
plt.plot(df_spx_ftse["new_date"],df_spx_ftse["GSPC500"])
plt.plot(df_spx_ftse["new_date"],df_spx_ftse["FTSE100"])
plt.title("S&P vs FTSE returns (2000 - 2010)", fontsize = 14, fontweight = 'bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

```python
df_spx_ftse_half2_2008 = df_spx_ftse[(df_spx_ftse.new_date >= '2008-07-01') &
                             (df_spx_ftse.new_date <= '2008-12-31')]
df_spx_ftse_half2_2008
```

```python
#To clearly distinguish between lines and observe patterns in the data we increase the granularity of the chart choosing a smaller timeframe
#labels = ["S&P", "FTSE 100"]
plt.figure(figsize = (20,8))
plt.plot(df_spx_ftse_half2_2008["new_date"],df_spx_ftse_half2_2008["GSPC500"], color = "midnightblue")
plt.plot(df_spx_ftse_half2_2008["new_date"],df_spx_ftse_half2_2008["FTSE100"], color = "crimson")
plt.title("S&P vs FTSE returns (The second half of 2008)", fontsize = 14, fontweight = 'bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

## Histogram

```python
df_real_estate = pd.read_csv(r"data/histogram_data.csv")
df_real_estate
```

```python
#The distribution of the price column
sns.set_style("white")
plt.figure(figsize = (8,6))
plt.hist(df_real_estate["Price"],
         bins = 8,
         color = "#108A99")
plt.title("Distribution of Real Estate prices", fontsize = 14, fontweight = 'bold')
plt.xlabel("Price in (000' $)")
plt.ylabel("Number of properties")
sns.despine()
plt.show()
```

## Scatter plot

```python
df_real_estate_data = pd.read_csv(r"data/scatter_data.csv")
df_real_estate_data
```

```python
sns.set_style("darkgrid")
plt.figure(figsize = (12,8))
scatter = plt.scatter(df_real_estate_data["Area (ft.)"],
            df_real_estate_data["Price"],
            alpha = 0.6,
            c = df_real_estate_data["Building Type"],
            cmap = 'viridis')
plt.legend(*scatter.legend_elements(), loc = 'upper left', title = 'Building Type')
plt.title("Relationship between Area and Price of California Real Estate", fontsize = 14, fontweight = 'bold')
plt.xlabel("Area (ft.)", weight  = 'bold')
plt.ylabel("Price in (000' $)", weight  = 'bold')
plt.show()
```

```python
#Option number 2 for creating a scatter plot
sns.set_style("darkgrid")
plt.figure(figsize = (12,8))
sns.scatterplot(df_real_estate_data["Price"],
                df_real_estate_data["Area (ft.)"],
                hue = df_real_estate_data["Building Type"],
                palette = ['black','darkblue','purple','pink','white'],
                s = 100)
plt.title("Relationship between Area and Price of California Real Estate", fontsize = 14, fontweight = 'bold')
plt.xlabel("Area (ft.)", weight  = 'bold')
plt.ylabel("Price in (000' $)", weight  = 'bold')
plt.show()
```
